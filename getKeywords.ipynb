# Per-chapter Keywords from Edward T. Hall's "Beyond Culture"


## Keyword extraction test
What we learn here is that this might work reasonably well on short texts, but starts to fall apart on longer ones.

Chapter seem to work but we need to be choosy. Nothing below a certain TextRank ranking.

Interesting experiement would be to do this on a book, with keywords teased out into the left or right margin of each paragraph.



```python
import os
import spacy
import pytextrank

path = 'chapters/'
files = sorted(os.listdir(path))

for f in files:
    print("\n------------\n")
    print("Chapter " + f[0:2] + "\n")
    with open(path+f, 'r') as file:
        text = file.read().replace('\n', '')

    # load a spaCy model, depending on language, scale, etc.
    nlp = spacy.load("en_core_web_sm")

    # add PyTextRank to the spaCy pipeline
    tr = pytextrank.TextRank()
    nlp.add_pipe(tr.PipelineComponent, name="textrank", last=True)

    doc = nlp(text)

    # examine the top-ranked phrases in the document
    for p in doc._.phrases:
        #print("{:.4f} {:5d}  {}".format(p.rank, p.count, p.text))
        #print(p.chunks)
        #print("\n")

        wc = p.text.split()
        is_between = 1 <= len(wc) <= 3
        if p.rank > 0.045:
            good_rank = True
        else:
            good_rank = False
        if is_between and good_rank:
            #print(p.rank)
            #print("\t—\t\t" + p.text)
            print(p.text)

```
